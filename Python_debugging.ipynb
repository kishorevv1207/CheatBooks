## Question 1

```python
def func(item, list_=[]):
     list_.append(item)
     return list_

x=func(1)
x=func(2)
x=func(3,[])
x=func(4)
print(x)
```

### Explanation

###### Options:

1. [4]
2. [1, 2, 3, 4]
3. [3, 4]
4. [1, 2, 4]

#### Answer:

[1, 2, 4]

Default arguments are evaluated only ONCE, at function definition time, not each time the function is called.


So list_=[] is created one time, and the same list is reused across calls unless you pass your own list.



Step-by-Step Execution

First call:

 - x = func(1)

 - list_ refers to the default list → []

 - Append 1

 - List becomes: [1]



Second call:

 - x = func(2)

 - list_ is the same default list (not a new one!)

 - Append 2

 - List becomes: [1, 2]





Third call (important difference):

 - x = func(3, [])

 - You explicitly pass a new empty list

 - Append 3

 - This list is temporary and not reused



Fourth call:

 - x = func(4)

 - Back to the original default list It was [1, 2]

 - Append 4

 - Final Output [1, 2, 4]





DEFAULT LIST (shared):

 - [] → [1] → [1, 2] → [1, 2, 4]



TEMP LIST (one-time):

 - [] → [3]



Why Python Does This



 - Default values are stored in function memory for performance

 - Mutable objects (like lists, dicts) keep their state






## Question 2

```python
x = float('nan')
print(x == x)
print(x is x)
```

### Explanation

#### Options:

1. True True
2. True False
3. False True
4. False False


#### Answer

- float('nan') creates a special floating-point value: NaN (Not a Number).

- By definition in IEEE 754 floating-point arithmetic:

- NaN is not equal to anything, including itself.

- So x == x → False.

- But in Python:

- x is x it checks object identity, not value equality.

- x is x → True

## Question 4

```python
def cal(a,b):
  return a+b
  return a-b

def cal(a,b):
  return a-b
  return a+b

print(cal(10,5))
```

### Explanation

#### Options

- 5
- 15
- SyntaxError:function redefined
- (5,15)

#### Answers:

The second definition of cal overwrites the first. Inside that function, return a-b executes immediately, so 10-5 = 5. The second return is unreachable.

## Question 9

```python
a = 1
b = 2
if a == 1 and b == 2:
  print(True)
else:
  print(False)

if a == 1 & b == 2:
  print(True)
else:
  print(False)
```

### Explanation

#### Options

- True True
- True False
- False True
- False False

#### Answers:

- & is NOT the same as and.
- & is a bitwise operator, not a logical operator.

Python evaluates it like this:

Step-by-step:

- 1 & b → 1 & 2
- Binary of 1 → 01
- Binary of 2 → 10
- 01 & 10 → 00 → value is 0

Now the condition becomes:

- a == 1 & b == 2:
- a == 0 == 2

This is False
